#### Import dependencies

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np
```

#### Create new csv for combined data

```python
files = [file for file in os.listdir('qb_2020_sales_item/months')]

all_months_data = pd.DataFrame()
months_file_order = [4,2,1,6,3,5]
month_order_count = 0
for file in files:
    df = pd.read_csv('qb_2020_sales_item/months/' + file,thousands=',')
    df['Month'] = months_file_order[month_order_count]
    month_order_count += 1
    all_months_data = pd.concat([all_months_data,df])
all_months_data.to_csv('mnm_all_months_2020_qb_sales_item.csv')
```

#### What month had the largest sales?

```python
sum_by_month = all_months_data.groupby('Month').sum()
```

```python
x = range(1,7)
y = sum_by_month['Amount']
plt.bar(x,y)
plt.xticks(x)
plt.xlabel('Months')
plt.ylabel('$ Amount')
plt.title('2020 Sales Amount by Month')
plt.show()
```

#### Drop all rows not inventory related

```python
inventory_df = df.drop(list_drop)
```

```python
no_totals_df = inventory_df[inventory_df['Inventory'].str[0:5] != 'Total']
no_totals_df.reset_index(inplace=True,drop=True)
no_totals_df
```

#### Inventory List

```python
full_inventory_list = no_totals_df['Inventory'].fillna(0)
full_inventory_list
```

```python
j=-1
for i in range(len(full_inventory_list)):
    if full_inventory_list[i] == 0:
        full_inventory_list[i] = inventory_list[j]
    else:
        j += 1
```
